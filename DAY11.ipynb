# STRING PALINDROME

```python
STR = str(input("enter string"))

A = STR[::-1]

print(A)

if STR == A:
    print("pailndrome")
else:
    print("Not Palindrome")
```

## FOR LOOP

```python
STR = str(input("enter string"))
A = ""
for i in STR:
    A = i + A
if STR == A:
    print("palindrome")
else:
    print("Not Plaindrome")
```
