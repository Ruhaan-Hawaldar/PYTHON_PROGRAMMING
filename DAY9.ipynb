# Factorial of a Number : Iterative and Recursive

## ITERATIVE

```python
N = int(input("Enter number:"))

fact = 1

for i in range(1,N+1):
    fact = fact * i

print(fact)
```

## RECURSIVE

```python
N = int(input("Enter number:"))

def factorial(N):
    if N == 0:
        return 1

    return N * factorial(N-1)

print(factorial(N))
```
